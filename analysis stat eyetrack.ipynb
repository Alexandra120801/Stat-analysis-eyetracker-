```python
import pandas as pd
import os

# Путь к папке с файлами
folder_path = 'C:\\Users\\Sasha\\excel_final\\'

# Получаем список файлов в папке
data_files = [file for file in os.listdir(folder_path) if file.endswith('.xlsx') or file.endswith('.csv')]

# Инициализируем пустой датафрейм
combined_df = pd.DataFrame()

# Читаем каждый файл и объединяем данные
for file in data_files:
    file_path = os.path.join(folder_path, file)
    print(f"Чтение файла: {file_path}")

    try:
        # Проверка наличия данных в файле перед чтением
        if os.path.getsize(file_path) > 0:
            if file.endswith('.xlsx'):
                df = pd.read_excel(file_path)
            elif file.endswith('.csv'):
                df = pd.read_csv(file_path)
            
            combined_df = combined_df.append(df, ignore_index=True)
        else:
            print(f"Файл {file} пустой. Пропуск.")

    except Exception as e:
        print(f"Произошла ошибка при чтении файла {file}: {e}")

# Сохраняем объединенные данные в CSV файл
combined_df.to_csv('combined_data.csv', index=False)
```

```python
# Функция для преобразования значений в столбце task
def transform_task(task_value):
    if isinstance(task_value, float) and pd.notna(task_value):
        return str(int(task_value))
    elif isinstance(task_value, str):
        if task_value.startswith('1'):
            return '1'
        elif task_value.startswith('2'):
            return '2'
        elif task_value.startswith('3'):
            return '3'
    return task_value
df['task'] = df['task'].astype(str).apply(transform_task)
```

```python
sample_size = 181
sampled_df = pd.DataFrame()

for task_type in df['task'].unique():

    sampled_data = df[df['task'] == task_type].sample(sample_size, random_state=42)
    sampled_df = pd.concat([sampled_df, sampled_data])

print(sampled_df['task'].value_counts())
```

```python
from scipy.stats import kruskal
columns_to_compare = ['click1', 'click2', 'menus', 'time','mistakes']

# Уровень значимости после коррекции методом Бонферрони
alpha_corrected = 0.05 / len(columns_to_compare)

# Перебираем все столбцы
for column in columns_to_compare:
    print(f"Сравнение для столбца {column}:")
    data_by_task = [sampled_df[sampled_df['task'] == task][column] for task in range(1, 4)]
    if all(data.notnull().any() for data in data_by_task):
        # Проведение теста Краскела-Уоллиса
        stat, p_value = kruskal(*data_by_task)
        print(f"Тест Краскела-Уоллиса: Statistic={stat}, p-value={p_value}")

        # Проверка статистической значимости после коррекции методом Бонферрони
        if p_value < alpha_corrected:
            print(f"Различия статистически значимы (после коррекции)")
        else:
            print("Различия не статистически значимы (после коррекции)")
    else:
        print("Данные отсутствуют в одной из групп")

    print("\n")
```

```python
from scipy.stats import mannwhitneyu

def mann_whitneyu_test(df, group_column, variable_column):
    groups = df[group_column].unique()
    results = []

    for i in range(len(groups)):
        for j in range(i + 1, len(groups)):
            group_i = groups[i]
            group_j = groups[j]

            data_i = df[df[group_column] == group_i][variable_column]
            data_j = df[df[group_column] == group_j][variable_column]

            stat, p_value = mannwhitneyu(data_i, data_j)
            result = {
                'Group I': group_i,
                'Group J': group_j,
                'Statistic': stat,
                'P-Value': p_value
            }
            results.append(result)

    return pd.DataFrame(results)

# Попарные сравнения для 'time' с использованием теста Манна-Уитни
mannwhitneyu_time = mann_whitneyu_test(df, 'task', 'time')
print("Pairwise Mann-Whitney U test for 'time':")
print(mannwhitneyu_time)

# Попарные сравнения для 'mistakes' с использованием теста Манна-Уитни
mannwhitneyu_mistakes = mann_whitneyu_test(df, 'task', 'mistakes')
print("\nPairwise Mann-Whitney U test for 'mistakes':")
print(mannwhitneyu_mistakes)
```
